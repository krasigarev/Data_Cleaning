# Merge, Join and Concatenate DataFrame

![merge_join_concatenate.jpg](data/merge_join_concatenate.jpg)

## Importing Libraries

```python
from os.path import join

import pandas as pd
```

## Loading Data

```python
df1 = pd.read_csv('./data/lotr_1.csv')
df1
```

```python
df2 = pd.read_csv('./data/lotr_2.csv')
df2
```

## Merge DataFrame

```python
# Default how='inner'
df1.merge(df2)
```

```python
# With how='inner'
df1.merge(df2, how='inner')
```

```python
# With how='inner'
df1.merge(df2, how='inner', on='FellowshipID')
```

```python
# With how='inner'
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
# With how='outer'
df1.merge(df2, how='outer')
```

```python
# With how='left'
df1.merge(df2, how='left')
```

```python
# With how='right'
df1.merge(df2, how='right')
```

```python
# With how='cross'
df1.merge(df2, how='cross')
```

## Join DataFrame

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_left', rsuffix='_right')
```

### Set index column

Example:

**df1.set_index**('Column_Name').join(**df2.set_index**('Column_Name')

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_left', rsuffix='_right', how='outer')
df4
```

```python
df5 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how='left', lsuffix='_left', rsuffix='_right')
df5
```

```python
df6 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how='right', lsuffix='_left', rsuffix='_right')
df6
```

## Concatenate DataFrame

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join='inner')
```

```python
pd.concat([df1, df2], join='outer', axis=1)
```
