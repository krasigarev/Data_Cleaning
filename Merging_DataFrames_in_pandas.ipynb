## Importing Libraries

```python
import pandas as pd
```

## Loading Data

```python
df1 = pd.read_csv('./data/lotr_1.csv')
df1
```

```python
df2 = pd.read_csv('./data/lotr_2.csv')
df2
```

## Merge DataFrame

```python
# Default how='inner'
df1.merge(df2)
```

```python
# With how='inner'
df1.merge(df2, how='inner')
```

```python
# With how='inner'
df1.merge(df2, how='inner', on='FellowshipID')
```

```python
# With how='inner'
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
# With how='outer'
df1.merge(df2, how='outer')
```
