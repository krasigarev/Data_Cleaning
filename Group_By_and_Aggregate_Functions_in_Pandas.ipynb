## Group By and Aggregate Functions in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/Flavor.csv')
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'sum', 'min', 'max', 'count']})
```

```python
df.groupby('Base Flavor').agg({
                                'Flavor Rating': ['mean', 'sum', 'min', 'max', 'count'],
                                'Texture Rating': ['mean', 'sum', 'min', 'max', 'count']
                              })
```

```python
df
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'sum', 'min', 'max', 'count']})
```

```python
df.groupby('Base Flavor').describe()
```

```python
df.groupby('Base Flavor').describe().T
```
