```python
import pandas as pd
```

## Importing Data

```python
df = pd.read_excel('./data/Customers.xlsx')
df
```

## Duplicates data

```python
df = df.drop_duplicates()
df
```

## Drop column

```python
df = df.drop(columns=['Not_Useful_Column'])
df
```

## Cleaning Last Name version 1

```python
df['Last_Name'] = df['Last_Name'].str.strip()
df['Last_Name'] = df['Last_Name'].str.lstrip("...")
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df
```

## Cleaning Last Name version 2

```python
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df
```

## Cleaning Phone_Number

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
df
```

## Cleaning Address

```python
df[["Street Address", "State", "Zip_Code"]] = df['Address'].str.split(',', expand=True)
df
```

## Clean Column Paying Customer

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
df
```

## Clean Column Do Not Contact

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

## Remove N/a and NaN

```python
# N/a
df = df.replace('N/a', '')
df
```

```python
# NaN
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
df['Do_Not_Contact'] = df["Do_Not_Contact"].replace('', 'N')
df
```

```python
# Clean N/a version 1
# drop null values
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```

```python
# Clean N/a version 2
# drop null values
df = df.dropna(subset=['Phone_Number'], inplace=True)
df
```
